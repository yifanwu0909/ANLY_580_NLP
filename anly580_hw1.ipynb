### 1.6 Produce a dispersion plot of the four main protagonists in Sense and Sensibility: Elinor, Marianne, Edward, and Willoughby. What can you observe about the different roles played by the males and females in this novel? Can you identify the couples?

```python
text2.dispersion_plot(["Elinor", "Marianne", "Edward", "Willoughby"])
```

From the above dispersion plot we can see that, both of the female roles(Elinor and Marianne) appear more frequently and consistant than male roles(Edward and Willoughby) in this novel. Both of the female roles appears more uniformly through out the whole text. Male roles appear periodically through out the whole text, and show an almost complementary pattern between the two male names. Female roles are the predominant roles in Sense and Sensibility. Marianne and Willoughby are more likely to be couples because through the whole book, where Marianne seems to appear more frequently, Willoughby also appears more frequently, although not as frequent as Marianne. This is an indication that they might be a couple. 

### 1.22  Find all the four-letter words in the Chat Corpus (text5). With the help of a frequency distribution (FreqDist), show these words in decreasing order of frequency. 

```python
four_letters = [i for i in text5 if i.isalpha() and len(i) == 4] #Filter out all 4-letters tokens and filter out anything that contain non-alphabetical characters.
FreqDist(four_letters).most_common()
```

### 1.24 Write expressions for finding all words in text6 that meet the conditions listed below. The result should be in the form of a list of words: ['word1', 'word2', ...].

(a). Ending in ize

(b). Containing the letter z

(c). Containing the sequence of letters pt

(d). Having all lowercase letters except for an initial capital (i.e., titlecase)


```python
#(a). 
[w for w in set(text6) if w.endswith('ize')]
```

```python
#(b).
sorted([w for w in set(text6) if 'z' in w])
```

```python
#(c).
sorted([w for w in set(text6) if 'pt' in w])
```

```python
#(d).
sorted([w for w in set(text6) if w.istitle()])
```

### 1.28 Define a function percent(word, text) that calculates how often a given word occurs in a text, and expresses the result as a percentage.

```python
def percent(word, text):
    return 100 * text.count(word) / len(text)
```
